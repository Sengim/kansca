```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import kan
```

```python
kan.utils.add_symbolic('-x^2', lambda x: -x**2)
```

```python
fig = plt.figure(figsize=(5,50))
x = torch.tensor(np.arange(-2*np.pi, 2*np.pi, 0.01))
n = len(kan.SYMBOLIC_LIB.keys())
for i, k in enumerate(kan.SYMBOLIC_LIB.keys()):
    ax = fig.add_subplot(n,1,i+1)
    ax.set_title(k)
    ax.plot(x, kan.SYMBOLIC_LIB[k][0](x).numpy())
plt.tight_layout()
plt.show()
```
